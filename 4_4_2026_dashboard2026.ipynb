```python
# ============================================================
# Resource Allocation / Maintenance — Final Version WITH K-Fold
# FULL MODULE (CLEAN + COMPLETE + SUBMISSION-READY)
# ============================================================

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import (
    train_test_split,
    StratifiedKFold,
    KFold,
    cross_val_score
)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# ============================================================
# 0) CONFIG
# ============================================================

DATA_PATH = "/content/maintence_dataset_final_hijri_added_and0prio.. .xlsx"
RANDOM_STATE = 42

TARGET_PRIORITY  = "priority"
TARGET_TECHCOUNT = "technicians_count"
TARGET_TRUCKS    = "no_of_cars"
TARGET_SPARE     = "need_spare_parts"

TEXT_COL = "fault_alarmandactivity"

BASE_CAT_COLS = ["site", "severity_category"]
BASE_NUM_COLS = ["no_technicians_at_time_of_maintance"]

PRIORITY_NUM_COLS = [
    "hijri_day", "hijri_month", "hijri_year",
    "incident_time",
    "risk_before_%",
    "distance_to_site_km",
]

PRIORITY_CAT_COLS = [
    "site", "severity_category",
    "days_type", "team_location", "sector",
    "maintenance_type", "responsible_team", "under_warranty"
]

TOP_EDA_PLOTS = [
    "hijri_day", "hijri_month", "hijri_year", "priority",
    "no_technicians_at_time_of_maintance",
    "no_of_cars", "technicians_count",
    "incident_time", "distance_to_site_km",
    "risk_before_%"
]

# ============================================================
# 1) LOAD + CLEAN
# ============================================================

def clean_column_name(c: str) -> str:
    c = str(c).strip().replace("\n", " ")
    while "  " in c:
        c = c.replace("  ", " ")
    return c

def resolve_tech_at_time_column(df: pd.DataFrame) -> pd.DataFrame:
    a = "no_technicians_at_time_of_maintance"
    b = "no_technicians_at_time_of_maintenance"
    if a not in df.columns and b in df.columns:
        df = df.rename(columns={b: a})
    return df

df = pd.read_excel(DATA_PATH)
df = df.rename(columns={c: clean_column_name(c) for c in df.columns})
df = resolve_tech_at_time_column(df)
df = df.drop_duplicates()

print("✅ Loaded:", DATA_PATH)
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

if TEXT_COL not in df.columns:
    raise ValueError(f"Missing required text column: {TEXT_COL}")

BASE_CAT_COLS = [c for c in BASE_CAT_COLS if c in df.columns]
BASE_NUM_COLS = [c for c in BASE_NUM_COLS if c in df.columns]
PRIORITY_NUM_COLS = [c for c in PRIORITY_NUM_COLS if c in df.columns]
PRIORITY_CAT_COLS = [c for c in PRIORITY_CAT_COLS if c in df.columns]

print("\n✅ Using Base Features (non-priority models):")
print("BASE_NUM_COLS:", BASE_NUM_COLS)
print("BASE_CAT_COLS:", BASE_CAT_COLS)
print("TEXT_COL     :", TEXT_COL)

print("\n✅ Using Priority Features:")
print("PRIORITY_NUM_COLS:", PRIORITY_NUM_COLS)
print("PRIORITY_CAT_COLS:", PRIORITY_CAT_COLS)
print("TEXT_COL         :", TEXT_COL)

display(df.head())

# ============================================================
# 2) DATASET SUMMARY
# ============================================================

eda_summary = pd.DataFrame({
    "Column": df.columns,
    "Data Type": df.dtypes.astype(str),
    "Missing Values": df.isna().sum(),
    "Missing %": (df.isna().mean() * 100).round(2),
    "Unique Values": df.nunique()
}).reset_index(drop=True)

print("\n✅ Dataset Summary Table")
display(eda_summary)

# ============================================================
# 3) QUICK EDA PLOTS
# ============================================================

def _to_numeric(series):
    return pd.to_numeric(series, errors="coerce")

def plot_hist(ax, s, title, bins=30):
    s = _to_numeric(pd.Series(s)).dropna()
    if len(s) == 0:
        ax.axis("off")
        return
    ax.hist(s, bins=bins, linewidth=0.1, edgecolor="black")
    ax.set_title(title, fontsize=9)
    ax.tick_params(axis="both", labelsize=8)

def plot_bar_counts(ax, s, title, rotate=45):
    s = pd.Series(s).dropna()
    s_num = pd.to_numeric(s, errors="coerce")

    if s_num.notna().sum() > 0:
        counts = s_num.dropna().value_counts().sort_index()
        labels = counts.index.astype(str).tolist()
        values = counts.values
    else:
        counts = s.astype(str).value_counts().head(20)
        labels = counts.index.astype(str).tolist()
        values = counts.values

    ax.bar(range(len(labels)), values, width=0.5, linewidth=0.4, edgecolor="black")
    ax.set_title(title, fontsize=9)
    ax.tick_params(axis="both", labelsize=8)
    ax.set_xticks(range(len(labels)))
    ax.set_xticklabels(labels, rotation=rotate, ha="right")

cols_exist = [c for c in TOP_EDA_PLOTS if c in df.columns]
n = len(cols_exist)

rows, cols = 3, 4
fig, axes = plt.subplots(rows, cols, figsize=(14, 7))
plt.suptitle("EDA Snapshot", fontsize=14)

k = 0
for r in range(rows):
    for c in range(cols):
        ax = axes[r, c]
        if k >= n:
            ax.axis("off")
            continue
        col = cols_exist[k]
        unique_vals = df[col].nunique(dropna=True)
        if unique_vals <= 15:
            plot_bar_counts(ax, df[col], col)
        else:
            plot_hist(ax, df[col], col)
        k += 1

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# ============================================================
# 4) PREPROCESS FACTORY
# ============================================================

def _extract_first_float(v: str):
    if v is None:
        return np.nan
    s = str(v).lower().strip().replace(",", "")
    matches = re.findall(r"-?\d+\.?\d*", s)
    return float(matches[0]) if matches else np.nan

def coerce_numeric_2d(X):
    X = pd.DataFrame(X).copy()
    for col in X.columns:
        X[col] = X[col].apply(_extract_first_float)
    return X

def make_preprocess(num_cols, cat_cols, text_col):
    num_pipe = Pipeline(steps=[
        ("to_num", FunctionTransformer(coerce_numeric_2d, validate=False)),
        ("imputer", SimpleImputer(strategy="median")),
    ])

    cat_pipe = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("ohe", OneHotEncoder(handle_unknown="ignore")),
    ])

    txt_pipe = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("to_1d", FunctionTransformer(lambda x: np.ravel(x).astype(str), validate=False)),
        ("tfidf", TfidfVectorizer(
            max_features=12000,
            ngram_range=(1, 2),
            min_df=2,
            max_df=0.90,
            sublinear_tf=True
        )),
    ])

    return ColumnTransformer(
        transformers=[
            ("num", num_pipe, num_cols),
            ("cat", cat_pipe, cat_cols),
            ("txt", txt_pipe, [text_col]),
        ],
        remainder="drop"
    )

print("\n✅ Preprocess factory ready")

# ============================================================
# 5) BUILD TARGETS
# ============================================================

df["y_priority"]   = pd.to_numeric(df.get(TARGET_PRIORITY), errors="coerce")
df["y_techcount"]  = pd.to_numeric(df.get(TARGET_TECHCOUNT), errors="coerce")
df["y_trucks"]     = pd.to_numeric(df.get(TARGET_TRUCKS), errors="coerce")

if TARGET_SPARE in df.columns:
    df["y_spare"] = df[TARGET_SPARE].astype(str).str.strip().str.lower()
    df.loc[df["y_spare"].isin(["nan", "none", ""]), "y_spare"] = np.nan
else:
    df["y_spare"] = np.nan

print("\nMissing y_priority :", round(df["y_priority"].isna().mean(), 3))
print("Missing y_techcount:", round(df["y_techcount"].isna().mean(), 3))
print("Missing y_trucks   :", round(df["y_trucks"].isna().mean(), 3))
print("Missing y_spare    :", round(df["y_spare"].isna().mean(), 3))

# ============================================================
# 6) TRAIN COLS
# ============================================================

PRIORITY_TRAIN_COLS = PRIORITY_NUM_COLS + PRIORITY_CAT_COLS + [TEXT_COL]
GENERAL_TRAIN_COLS  = BASE_NUM_COLS + BASE_CAT_COLS + [TEXT_COL]

print("\n✅ Priority TRAIN COLS (Priority=1..4 only):", PRIORITY_TRAIN_COLS)
print("✅ General TRAIN COLS:", GENERAL_TRAIN_COLS)

# ============================================================
# 7) MODEL BUILDERS
# ============================================================

def build_priority_model():
    return Pipeline(steps=[
        ("prep", make_preprocess(PRIORITY_NUM_COLS, PRIORITY_CAT_COLS, TEXT_COL)),
        ("model", LinearSVC(C=2.0, class_weight="balanced"))
    ])

def build_spare_model():
    return Pipeline(steps=[
        ("prep", make_preprocess(BASE_NUM_COLS, BASE_CAT_COLS, TEXT_COL)),
        ("model", LogisticRegression(
            max_iter=5000,
            n_jobs=-1,
            solver="saga",
            class_weight="balanced"
        ))
    ])

def build_tech_model():
    return Pipeline(steps=[
        ("prep", make_preprocess(BASE_NUM_COLS, BASE_CAT_COLS, TEXT_COL)),
        ("model", RandomForestRegressor(
            n_estimators=500,
            random_state=RANDOM_STATE,
            n_jobs=-1,
            min_samples_leaf=2
        ))
    ])

def build_trucks_model():
    return Pipeline(steps=[
        ("prep", make_preprocess(BASE_NUM_COLS, BASE_CAT_COLS, TEXT_COL)),
        ("model", RandomForestRegressor(
            n_estimators=500,
            random_state=RANDOM_STATE,
            n_jobs=-1,
            min_samples_leaf=2
        ))
    ])

# ============================================================
# 8) K-FOLD HELPERS
# ============================================================

def _safe_stratified_kfold(y, n_splits=5, random_state=42):
    vc = pd.Series(y).value_counts()
    min_class = int(vc.min()) if len(vc) > 0 else 0
    k = min(n_splits, min_class) if min_class > 1 else 2
    k = max(2, k)
    return StratifiedKFold(n_splits=k, shuffle=True, random_state=random_state), k

def kfold_eval_classification(X, y, build_model_fn, label_name="target", n_splits=5):
    skf, k_used = _safe_stratified_kfold(y, n_splits=n_splits, random_state=RANDOM_STATE)
    accs, f1s = [], []

    for fold, (tr, te) in enumerate(skf.split(X, y), 1):
        model = build_model_fn()
        model.fit(X.iloc[tr], y.iloc[tr])
        pred = model.predict(X.iloc[te])
        accs.append(accuracy_score(y.iloc[te], pred))
        f1s.append(f1_score(y.iloc[te], pred, average="macro"))

    return {
        "Task": f"KFold Classification ({label_name})",
        "FoldsUsed": k_used,
        "Accuracy_mean": float(np.mean(accs)),
        "Accuracy_std": float(np.std(accs)),
        "MacroF1_mean": float(np.mean(f1s)),
        "MacroF1_std": float(np.std(f1s)),
    }

def kfold_eval_regression(X, y, build_model_fn, label_name="target", n_splits=5):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=RANDOM_STATE)
    maes, rmses, r2s = [], [], []

    for fold, (tr, te) in enumerate(kf.split(X), 1):
        model = build_model_fn()
        model.fit(X.iloc[tr], y.iloc[tr])
        pred = model.predict(X.iloc[te])
        maes.append(mean_absolute_error(y.iloc[te], pred))
        rmses.append(float(np.sqrt(mean_squared_error(y.iloc[te], pred))))
        r2s.append(r2_score(y.iloc[te], pred))

    return {
        "Task": f"KFold Regression ({label_name})",
        "FoldsUsed": n_splits,
        "MAE_mean": float(np.mean(maes)),
        "MAE_std": float(np.std(maes)),
        "RMSE_mean": float(np.mean(rmses)),
        "RMSE_std": float(np.std(rmses)),
        "R2_mean": float(np.mean(r2s)),
        "R2_std": float(np.std(r2s)),
    }

# ============================================================
# 9) K-FOLD DATA
# ============================================================

df_p_k = df.dropna(subset=["y_priority"]).copy()
df_p_k = df_p_k[df_p_k["y_priority"].between(1, 4)].copy()
Xp_k = df_p_k[PRIORITY_TRAIN_COLS]
yp_k = df_p_k["y_priority"].astype(int)

df_s_k = df.dropna(subset=["y_spare"]).copy()
Xs_k = df_s_k[GENERAL_TRAIN_COLS]
ys_k = df_s_k["y_spare"]

df_t_k = df.dropna(subset=["y_techcount"]).copy()
Xt_k = df_t_k[GENERAL_TRAIN_COLS]
yt_k = df_t_k["y_techcount"]

df_c_k = df.dropna(subset=["y_trucks"]).copy()
Xc_k = df_c_k[GENERAL_TRAIN_COLS]
yc_k = df_c_k["y_trucks"]

# ============================================================
# 10) RUN K-FOLD
# ============================================================

prio_kfold = kfold_eval_classification(
    X=Xp_k, y=yp_k,
    build_model_fn=build_priority_model,
    label_name="priority(1..4)",
    n_splits=5
)

spare_kfold = kfold_eval_classification(
    X=Xs_k, y=ys_k,
    build_model_fn=build_spare_model,
    label_name="need_spare_parts",
    n_splits=5
)

tech_kfold = kfold_eval_regression(
    X=Xt_k, y=yt_k,
    build_model_fn=build_tech_model,
    label_name="technicians_count",
    n_splits=5
)

trucks_kfold = kfold_eval_regression(
    X=Xc_k, y=yc_k,
    build_model_fn=build_trucks_model,
    label_name="no_of_cars",
    n_splits=5
)

kfold_results = pd.DataFrame([prio_kfold, spare_kfold, tech_kfold, trucks_kfold])
print("\n✅ K-Fold Cross-Validation Results")
display(kfold_results)

# ============================================================
# 11) HOLDOUT MODEL A — Priority
# ============================================================

data_p = df.dropna(subset=["y_priority"]).copy()
data_p = data_p[data_p["y_priority"].between(1, 4)].copy()
X_p = data_p[PRIORITY_TRAIN_COLS]
y_p = data_p["y_priority"].astype(int)

strat = y_p if y_p.value_counts().min() >= 2 else None

Xp_train, Xp_test, yp_train, yp_test = train_test_split(
    X_p, y_p, test_size=0.2, random_state=RANDOM_STATE, stratify=strat
)

priority_model = build_priority_model()
priority_model.fit(Xp_train, yp_train)
pred_p = priority_model.predict(Xp_test)

print("\n==============================")
print("✅ Priority Classification Results (1..4 only)")
print("==============================")
print("Accuracy :", round(accuracy_score(yp_test, pred_p), 3))
print("Macro-F1 :", round(f1_score(yp_test, pred_p, average='macro'), 3))
print(classification_report(yp_test, pred_p))

# ============================================================
# 12) HOLDOUT MODEL B — technicians_count
# ============================================================

data_t = df.dropna(subset=["y_techcount"]).copy()
X_t = data_t[GENERAL_TRAIN_COLS]
y_t = data_t["y_techcount"]

Xt_train, Xt_test, yt_train, yt_test = train_test_split(
    X_t, y_t, test_size=0.2, random_state=RANDOM_STATE
)

tech_count_model = build_tech_model()
tech_count_model.fit(Xt_train, yt_train)
pred_t = tech_count_model.predict(Xt_test)
rmse = float(np.sqrt(mean_squared_error(yt_test, pred_t)))

print("\n==============================")
print("✅ technicians_count Regression Results")
print("==============================")
print("MAE :", round(mean_absolute_error(yt_test, pred_t), 3))
print("RMSE:", round(rmse, 3))
print("R2  :", round(r2_score(yt_test, pred_t), 3))

# ============================================================
# 13) HOLDOUT MODEL C — no_of_cars
# ============================================================

data_c = df.dropna(subset=["y_trucks"]).copy()
X_c = data_c[GENERAL_TRAIN_COLS]
y_c = data_c["y_trucks"]

Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_c, y_c, test_size=0.2, random_state=RANDOM_STATE
)

trucks_model = build_trucks_model()
trucks_model.fit(Xc_train, yc_train)
pred_c = trucks_model.predict(Xc_test)
rmse_c = float(np.sqrt(mean_squared_error(yc_test, pred_c)))

print("\n==============================")
print("✅ no_of_cars Regression Results")
print("==============================")
print("MAE :", round(mean_absolute_error(yc_test, pred_c), 3))
print("RMSE:", round(rmse_c, 3))
print("R2  :", round(r2_score(yc_test, pred_c), 3))

# ============================================================
# 14) HOLDOUT MODEL D — need_spare_parts
# ============================================================

data_s = df.dropna(subset=["y_spare"]).copy()
X_s = data_s[GENERAL_TRAIN_COLS]
y_s = data_s["y_spare"]

strat_s = y_s if y_s.value_counts().min() >= 2 else None

Xs_train, Xs_test, ys_train, ys_test = train_test_split(
    X_s, y_s, test_size=0.2, random_state=RANDOM_STATE, stratify=strat_s
)

spare_model = build_spare_model()
spare_model.fit(Xs_train, ys_train)
pred_s = spare_model.predict(Xs_test)

print("\n==============================")
print("✅ need_spare_parts Classification Results")
print("==============================")
print("Accuracy :", round(accuracy_score(ys_test, pred_s), 3))
print("Macro-F1 :", round(f1_score(ys_test, pred_s, average='macro'), 3))
print(classification_report(ys_test, pred_s))

# ============================================================
# 15) RESULTS TABLE
# ============================================================

results_table = pd.DataFrame([
    {
        "Task": "Priority Classification (1..4 only)",
        "Target": TARGET_PRIORITY,
        "Model": "LinearSVC (balanced)",
        "Metrics": f"Accuracy={accuracy_score(yp_test, pred_p)*100:.1f}%, Macro-F1={f1_score(yp_test, pred_p, average='macro')*100:.1f}%"
    },
    {
        "Task": "Technicians Estimation",
        "Target": TARGET_TECHCOUNT,
        "Model": "RandomForestRegressor",
        "Metrics": f"MAE={mean_absolute_error(yt_test, pred_t):.3f}, RMSE={rmse:.3f}, R2={r2_score(yt_test, pred_t):.3f}"
    },
    {
        "Task": "Trucks/Cars Estimation",
        "Target": TARGET_TRUCKS,
        "Model": "RandomForestRegressor",
        "Metrics": f"MAE={mean_absolute_error(yc_test, pred_c):.3f}, RMSE={rmse_c:.3f}, R2={r2_score(yc_test, pred_c):.3f}"
    },
    {
        "Task": "Spare Parts Needed",
        "Target": TARGET_SPARE,
        "Model": "LogisticRegression (balanced saga)",
        "Metrics": f"Accuracy={accuracy_score(ys_test, pred_s)*100:.1f}%, Macro-F1={f1_score(ys_test, pred_s, average='macro')*100:.1f}%"
    }
])

print("\n✅ Model Evaluation Results")
display(results_table)

# ============================================================
# 16) POLICY CONSTRAINTS
# ============================================================

EMERGENCY_KEYWORDS = [
    "fire", "smoke", "gas leak", "power outage",
    "short circuit", "electric shock",
    "elevator stuck", "lift stuck",
    "explosion", "sparks",
]

EMERGENCY_MIN_TECH = 3
EMERGENCY_MIN_TRUCKS = 2
EMERGENCY_FORCE_SPARE = True

MAX_TECH_CAP = 6
MAX_TRUCKS_CAP = 3

def _has_emergency_keywords(text: str) -> bool:
    t = str(text).lower()
    return any(k in t for k in EMERGENCY_KEYWORDS)

def _to_float_safe(x):
    try:
        if x is None:
            return None
        return float(x)
    except Exception:
        return None

def _to_int_safe(x, default=None):
    try:
        return int(x)
    except Exception:
        return default

def _downtime_escalate_priority(pri: int, hours_downtime: float):
    hd = _to_float_safe(hours_downtime)
    if hd is None:
        return pri, None

    if hd >= 24:
        new_pri = 1
        if pri != new_pri:
            return new_pri, f"Downtime Escalation: hours_downtime({hd}) >= 24 -> priority {pri} -> {new_pri}"
        return pri, None

    if hd >= 10:
        new_pri = 2
        if pri > new_pri:
            return new_pri, f"Downtime Escalation: hours_downtime({hd}) >= 10 -> priority {pri} -> {new_pri}"
        return pri, None

    return pri, None

def apply_policy_constraints(input_row: dict, model_output: dict, hours_downtime: float = None):
    reasons = []

    pri = int(model_output["predicted_priority"])
    tech = int(model_output["predicted_technicians_count"])
    trucks = int(model_output["predicted_trucks_count"])
    spare = str(model_output["predicted_need_spare_parts"]).lower().strip()

    if spare in ["1", "true", "yes", "y", "needed", "need"]:
        spare = "yes"
    elif spare in ["0", "false", "no", "n", "none", "not needed"]:
        spare = "no"

    txt = str(input_row.get(TEXT_COL, ""))
    emergency = _has_emergency_keywords(txt)
    if emergency:
        reasons.append("Emergency Trigger: keyword detected in fault text")

    if emergency:
        if tech < EMERGENCY_MIN_TECH:
            reasons.append(f"Emergency Floors: technicians {tech} -> {EMERGENCY_MIN_TECH}")
            tech = EMERGENCY_MIN_TECH

        if trucks < EMERGENCY_MIN_TRUCKS:
            reasons.append(f"Emergency Floors: trucks {trucks} -> {EMERGENCY_MIN_TRUCKS}")
            trucks = EMERGENCY_MIN_TRUCKS

        if EMERGENCY_FORCE_SPARE and spare != "yes":
            reasons.append("Emergency Floors: spare parts forced to YES")
            spare = "yes"

    new_pri, pr_reason = _downtime_escalate_priority(pri, hours_downtime)
    if pr_reason is not None:
        reasons.append(pr_reason)
    pri = new_pri

    available = _to_int_safe(input_row.get("no_technicians_at_time_of_maintance"), None)
    if available is not None and tech > available:
        reasons.append(f"Availability Constraint: technicians {tech} -> {available} (available now)")
        tech = available

    if tech > MAX_TECH_CAP:
        reasons.append(f"Global Cap: technicians {tech} -> {MAX_TECH_CAP}")
        tech = MAX_TECH_CAP

    if trucks > MAX_TRUCKS_CAP:
        reasons.append(f"Global Cap: trucks {trucks} -> {MAX_TRUCKS_CAP}")
        trucks = MAX_TRUCKS_CAP

    final_output = dict(model_output)
    final_output.update({
        "final_priority": pri,
        "final_technicians_count": tech,
        "final_trucks_count": trucks,
        "final_need_spare_parts": spare
    })
    return final_output, reasons

# ============================================================
# 17) SAFE PREDICTION API + DEMO
# ============================================================

def predict_one(
    site,
    fault_alarmandactivity,
    severity_category,
    no_technicians_at_time_of_maintance,
    hours_downtime,
    day,
    month,
    year,
):
    site = str(site).lower().strip()
    severity_category = str(severity_category).lower().strip()
    alarm_text = str(fault_alarmandactivity)

    pr_row = {
        "hijri_day": day,
        "hijri_month": month,
        "hijri_year": year,
        "incident_time": None,
        "risk_before_%": None,
        "distance_to_site_km": None,
        "site": site,
        "severity_category": severity_category,
        "days_type": None,
        "team_location": None,
        "sector": None,
        "maintenance_type": None,
        "responsible_team": None,
        "under_warranty": None,
        TEXT_COL: alarm_text,
    }
    X_pr = pd.DataFrame([pr_row]).reindex(columns=PRIORITY_TRAIN_COLS)
    pri = int(priority_model.predict(X_pr)[0])

    gen_row = {
        "site": site,
        "severity_category": severity_category,
        TEXT_COL: alarm_text,
        "no_technicians_at_time_of_maintance": no_technicians_at_time_of_maintance,
    }
    X_gen = pd.DataFrame([gen_row]).reindex(columns=GENERAL_TRAIN_COLS)

    tech_raw = float(tech_count_model.predict(X_gen)[0])
    tech_final = int(round(max(1, tech_raw)))

    trucks_raw = float(trucks_model.predict(X_gen)[0])
    trucks_final = int(round(max(0, trucks_raw)))

    spare_pred = spare_model.predict(X_gen)[0]
    spare_conf = None
    if hasattr(spare_model.named_steps["model"], "predict_proba"):
        spare_conf = float(np.max(spare_model.predict_proba(X_gen)))

    model_output = {
        "predicted_priority": pri,
        "predicted_technicians_count": tech_final,
        "raw_technicians_count": tech_raw,
        "predicted_trucks_count": trucks_final,
        "raw_trucks_count": trucks_raw,
        "predicted_need_spare_parts": spare_pred,
        "spare_parts_confidence": spare_conf
    }

    final_output, reasons = apply_policy_constraints(
        input_row=gen_row,
        model_output=model_output,
        hours_downtime=hours_downtime
    )

    final_output["policy_reasons"] = reasons
    return final_output

def show_result(input_dict, output_dict):
    print("=" * 70)
    print("📥 INPUT (ONLY 8 fields)")
    print("-" * 70)
    for k, v in input_dict.items():
        print(f"{k:40s}: {v}")

    print("\n📤 MODEL OUTPUT (Before Policy)")
    print("-" * 70)
    print(f"{'Predicted Priority (1..4 only)':40s}: {output_dict['predicted_priority']}")
    print(f"{'Recommended Technicians Count':40s}: {output_dict['predicted_technicians_count']}  (raw={output_dict['raw_technicians_count']:.2f})")
    print(f"{'Recommended Trucks/Cars Count':40s}: {output_dict['predicted_trucks_count']}  (raw={output_dict['raw_trucks_count']:.2f})")

    if output_dict["spare_parts_confidence"] is not None:
        print(f"{'Need Spare Parts?':40s}: {output_dict['predicted_need_spare_parts']}  (conf={output_dict['spare_parts_confidence']:.3f})")
    else:
        print(f"{'Need Spare Parts?':40s}: {output_dict['predicted_need_spare_parts']}")

    print("\n📌 FINAL OUTPUT (After Policy Constraints)")
    print("-" * 70)
    print(f"{'FINAL Priority':40s}: {output_dict['final_priority']}")
    print(f"{'FINAL Technicians Count':40s}: {output_dict['final_technicians_count']}")
    print(f"{'FINAL Trucks/Cars Count':40s}: {output_dict['final_trucks_count']}")
    print(f"{'FINAL Need Spare Parts?':40s}: {output_dict['final_need_spare_parts']}")

    if output_dict.get("policy_reasons"):
        print("\n🧾 Policy Reasons")
        print("-" * 70)
        for r in output_dict["policy_reasons"]:
            print(f"- {r}")

    print("=" * 70)

input_data = {
    "site": "m010",
    "fault_alarmandactivity": "Power outage in main electrical panel",
    "severity_category": "major",
    "no_technicians_at_time_of_maintance": 2,
    "hours_downtime": 6,
    "day": 10,
    "month": 12,
    "year": 1446
}

output_data = predict_one(
    site=input_data["site"],
    fault_alarmandactivity=input_data["fault_alarmandactivity"],
    severity_category=input_data["severity_category"],
    no_technicians_at_time_of_maintance=input_data["no_technicians_at_time_of_maintance"],
    hours_downtime=input_data["hours_downtime"],
    day=input_data["day"],
    month=input_data["month"],
    year=input_data["year"],
)

show_result(input_data, output_data)
print("\nDone ✅")

# ============================================================
# 18) MODULE RELIABILITY
# ============================================================

def classify_reliability_label(cv_std, holdout_gap):
    if cv_std < 0.03 and holdout_gap < 0.03:
        return "HIGH"
    elif cv_std < 0.07 and holdout_gap < 0.07:
        return "MODERATE"
    else:
        return "LOW"

def regression_reliability_label(cv_std, holdout_gap):
    if cv_std < 0.05 and holdout_gap < 0.05:
        return "HIGH"
    elif cv_std < 0.10 and holdout_gap < 0.10:
        return "MODERATE"
    else:
        return "LOW"

def label_to_score(label):
    return {"HIGH": 3, "MODERATE": 2, "LOW": 1}[label]

def evaluate_priority_reliability(X, y):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=RANDOM_STATE,
        stratify=y
    )

    model = build_priority_model()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    holdout_f1 = f1_score(y_test, y_pred, average="macro")
    holdout_acc = accuracy_score(y_test, y_pred)

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    cv_f1_scores = cross_val_score(model, X, y, cv=cv, scoring="f1_macro")
    cv_acc_scores = cross_val_score(model, X, y, cv=cv, scoring="accuracy")

    f1_gap = abs(cv_f1_scores.mean() - holdout_f1)
    reliability = classify_reliability_label(cv_f1_scores.std(), f1_gap)

    return {
        "Output": TARGET_PRIORITY,
        "Type": "Classification",
        "CV_Main_Metric": "Macro-F1",
        "CV_Mean": round(cv_f1_scores.mean(), 4),
        "CV_STD": round(cv_f1_scores.std(), 4),
        "Holdout": round(holdout_f1, 4),
        "Gap": round(f1_gap, 4),
        "Extra_Metric": f"Accuracy CV={cv_acc_scores.mean():.4f}, Holdout={holdout_acc:.4f}",
        "Reliability_Label": reliability
    }

def evaluate_spare_reliability(X, y):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=RANDOM_STATE,
        stratify=y
    )

    model = build_spare_model()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    holdout_f1 = f1_score(y_test, y_pred, average="macro")
    holdout_acc = accuracy_score(y_test, y_pred)

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    cv_f1_scores = cross_val_score(model, X, y, cv=cv, scoring="f1_macro")
    cv_acc_scores = cross_val_score(model, X, y, cv=cv, scoring="accuracy")

    f1_gap = abs(cv_f1_scores.mean() - holdout_f1)
    reliability = classify_reliability_label(cv_f1_scores.std(), f1_gap)

    return {
        "Output": TARGET_SPARE,
        "Type": "Classification",
        "CV_Main_Metric": "Macro-F1",
        "CV_Mean": round(cv_f1_scores.mean(), 4),
        "CV_STD": round(cv_f1_scores.std(), 4),
        "Holdout": round(holdout_f1, 4),
        "Gap": round(f1_gap, 4),
        "Extra_Metric": f"Accuracy CV={cv_acc_scores.mean():.4f}, Holdout={holdout_acc:.4f}",
        "Reliability_Label": reliability
    }

def evaluate_tech_reliability(X, y):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=RANDOM_STATE
    )

    model = build_tech_model()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    holdout_r2 = r2_score(y_test, y_pred)
    holdout_mae = mean_absolute_error(y_test, y_pred)
    holdout_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    cv = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    cv_r2_scores = cross_val_score(model, X, y, cv=cv, scoring="r2")
    cv_mae_scores = -cross_val_score(model, X, y, cv=cv, scoring="neg_mean_absolute_error")

    r2_gap = abs(cv_r2_scores.mean() - holdout_r2)
    reliability = regression_reliability_label(cv_r2_scores.std(), r2_gap)

    return {
        "Output": TARGET_TECHCOUNT,
        "Type": "Regression",
        "CV_Main_Metric": "R2",
        "CV_Mean": round(cv_r2_scores.mean(), 4),
        "CV_STD": round(cv_r2_scores.std(), 4),
        "Holdout": round(holdout_r2, 4),
        "Gap": round(r2_gap, 4),
        "Extra_Metric": f"MAE CV={cv_mae_scores.mean():.4f}, Holdout={holdout_mae:.4f}, RMSE Holdout={holdout_rmse:.4f}",
        "Reliability_Label": reliability
    }

def evaluate_trucks_reliability(X, y):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.2,
        random_state=RANDOM_STATE
    )

    model = build_trucks_model()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    holdout_r2 = r2_score(y_test, y_pred)
    holdout_mae = mean_absolute_error(y_test, y_pred)
    holdout_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    cv = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    cv_r2_scores = cross_val_score(model, X, y, cv=cv, scoring="r2")
    cv_mae_scores = -cross_val_score(model, X, y, cv=cv, scoring="neg_mean_absolute_error")

    r2_gap = abs(cv_r2_scores.mean() - holdout_r2)
    reliability = regression_reliability_label(cv_r2_scores.std(), r2_gap)

    return {
        "Output": TARGET_TRUCKS,
        "Type": "Regression",
        "CV_Main_Metric": "R2",
        "CV_Mean": round(cv_r2_scores.mean(), 4),
        "CV_STD": round(cv_r2_scores.std(), 4),
        "Holdout": round(holdout_r2, 4),
        "Gap": round(r2_gap, 4),
        "Extra_Metric": f"MAE CV={cv_mae_scores.mean():.4f}, Holdout={holdout_mae:.4f}, RMSE Holdout={holdout_rmse:.4f}",
        "Reliability_Label": reliability
    }

priority_rel = evaluate_priority_reliability(Xp_k, yp_k)
spare_rel    = evaluate_spare_reliability(Xs_k, ys_k)
tech_rel     = evaluate_tech_reliability(Xt_k, yt_k)
trucks_rel   = evaluate_trucks_reliability(Xc_k, yc_k)

module_reliability_df = pd.DataFrame([
    priority_rel,
    spare_rel,
    tech_rel,
    trucks_rel
])

module_reliability_df["Reliability_Score"] = module_reliability_df["Reliability_Label"].map(label_to_score)
module_reliability_df["Reliability_Percentage"] = (
    module_reliability_df["Reliability_Score"] / 3 * 100
).round(2)

display(module_reliability_df)

overall_score = module_reliability_df["Reliability_Score"].mean()
overall_percentage = round((overall_score / 3) * 100, 2)

if overall_percentage >= 90:
    overall_module_reliability = "HIGH"
elif overall_percentage >= 75:
    overall_module_reliability = "MODERATE"
else:
    overall_module_reliability = "LOW"

print("\n===================================================")
print("Overall Module Reliability Label :", overall_module_reliability)
print("Overall Module Reliability Score :", round(overall_score, 2), "/ 3.00")
print("Overall Module Reliability (%)   :", overall_percentage, "%")
print("===================================================")
```
